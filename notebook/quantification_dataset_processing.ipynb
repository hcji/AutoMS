## Quantification dataset processing

### Import package

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from AutoMS.automs import AutoMS
```

### Data processing

```python
path = 'E:/Data/Quantification Dataset for KPIC2'
files = os.listdir(path)

output = {}
for f in tqdm(files):
    file = path + '/' + f
    peaks = AutoMS(file, min_intensity = 50000)
    output[f] = peaks
```

### Feature mathcing

```python
rtmzs = []
for i, peaks in output.items():
    for j in peaks.index:
        if peaks['score'][j] <= 0:
            continue
        mz = np.round(peaks['mz'][j], 2)
        rt = np.round(peaks['rt'][j], -1)
        rtmz = str(rt) + '_' + str(mz)
        if rtmz not in rtmzs:
            rtmzs.append(rtmz)
rts = np.array([float(m.split('_')[0]) for m in rtmzs])
mzs = np.array([float(m.split('_')[1]) for m in rtmzs])


intensities = np.full((len(rtmzs), len(output)), np.nan)
scores = np.full((len(rtmzs), len(output)), np.nan)
for i, peaks in output.items():
    for j in peaks.index:
        k1 = np.abs(peaks['mz'][j] - mzs) < 0.02
        k2 = np.abs(peaks['rt'][j] - rts) < 10
        kk = np.where(np.logical_and(k1, k2))[0]
        if len(kk) == 0:
            continue
        else:
            pass
        intensity = peaks['intensity'][j]
        score = peaks['score'][j]
        
        for k in kk:
            m = files.index(i)
            intensities[k, m] = np.nanmax([intensities[k, m], intensity])
            scores[k, m] = score
```

### Plot the linearity

```python
from scipy.stats import pearsonr

compounds = ['Niacin', 'Methionine', 'Phenylalanine', 'Hippuric acid', 'Tryptophan']
target_rt_mz = ['60.0_124.04', '70.0_150.06', '130.0_166.09', '220.0_180.07', '170.0_205.09']

plt.figure(dpi = 300, figsize=(14, 9))
for i, rtmz in enumerate(target_rt_mz):
    w = np.where(np.array(rtmzs) == rtmz)[0][0]
    abund = intensities[w, :]
    mean_abund = []
    concent = np.array([1,10,2,3,4,5,6,7,8,9])
    for j in range(10):
        mean_abund.append(np.nanmean(abund[(j)*6 : (1+j)*6]))
    plt.subplot(2,3,i+1)
    p = np.poly1d(np.polyfit(concent, mean_abund, 1))
    r2 = np.round(pearsonr(concent, mean_abund)[0], 4)
    plt.plot(concent, mean_abund, 'o')
    plt.plot(concent, p(concent), color = 'red')
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.text(1, 0.95*np.max(mean_abund), 'Corr={}'.format(r2), fontsize = 15)
    plt.title(label = compounds[i], loc = 'left', pad = 12, fontsize = 19, weight='bold')
    
plt.xlabel('Concentration (mg/L)', ha='center', fontsize = 22)
plt.text(-15.5, 9*10**6, 'Quantitative value', va='center', rotation='vertical', fontsize = 22)
```
